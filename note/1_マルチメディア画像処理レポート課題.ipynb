## 2. 画像を生成する
シェルピンスキーのギャスケットとカーペットを生成するプログラム合わせたクラスを作成する。

```python
import numpy as np

class makeImage:
    def __init__(self, w, h):
        self.w = w
        self.h = h
        self.pix = np.zeros(self.w*self.h)
        
    def main(self, argv):
        if argv == "t":
            self.setTriangle()
        elif argv == "s":
            self.setSquare()
        self.draw()
        
    def setTriangle(self):
        self.pix[int(self.w/2)] = 1
         
        for y in range(1,self.h):
            for x in range(1,self.w):
                p = y * (self.w) + x            
                self.pix[p] = self.pix[p-self.w-1] + self.pix[p-self.w+1]
                if self.pix[p] != 1:
                    self.pix[p] = 0
        
    def setSquare(self):
        self.pix[w/2] = 1
        cpt = np.zeros([3,3])
        sier_cpt = np.array([])
        counter = 1
        while self.w * self.h > 3**counter:
            for i in len(sier_cpt):
                sier_cpt.append(cp)
        
                
    def draw(self):
        # preprocess
        pix_2dim = []
        
        # 1*(w * h) vector => w * h vector
        for i in range(self.w):
            pix_2dim.append(self.pix[self.w*i:self.w*(i+1)])
        self.pix= pix_2dim
        
        plt.imshow(self.pix[:self.h], cmap='gray', interpolation='nearest', vmin=0, vmax=1)        
```

```python
mi = makeImage(400,200)
mi.main("t")
```
